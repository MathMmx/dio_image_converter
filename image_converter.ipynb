```python
import os
import requests
import numpy as np
from PIL import Image
from io import BytesIO
import random
#Utilizei imagens aleatorias do site pexels
imagens_urls = [
    'https://images.pexels.com/photos/1231234/pexels-photo-1231234.jpeg',
    'https://images.pexels.com/photos/2342345/pexels-photo-2342345.jpeg',
    'https://images.pexels.com/photos/3453456/pexels-photo-3453456.jpeg',
    'https://images.pexels.com/photos/4564567/pexels-photo-4564567.jpeg',
    'https://images.pexels.com/photos/5675678/pexels-photo-5675678.jpeg',
    'https://images.pexels.com/photos/6786789/pexels-photo-6786789.jpeg',
    'https://images.pexels.com/photos/7897890/pexels-photo-7897890.jpeg',
    'https://images.pexels.com/photos/8908901/pexels-photo-8908901.jpeg',
    'https://images.pexels.com/photos/9019012/pexels-photo-9019012.jpeg',
    'https://images.pexels.com/photos/0120123/pexels-photo-0120123.jpeg'
]

#Fiz uma funcao para baixar as imagens
def baixar_imagens(imagens_urls, pasta_entrada, n=10):
    if not os.path.exists(pasta_entrada):
        os.makedirs(pasta_entrada)

    imagens_selecionadas = random.sample(imagens_urls, n)

    for i, url_imagem in enumerate(imagens_selecionadas):
        try:
            resposta_imagem = requests.get(url_imagem)

            if resposta_imagem.status_code == 200:
                imagem = Image.open(BytesIO(resposta_imagem.content))
                caminho_imagem = os.path.join(pasta_entrada, f"imagem_{i+1}.jpg")
                imagem.save(caminho_imagem)
                print(f"Imagem {i+1} salva em {caminho_imagem}")
            else:
                print(f"Falha ao baixar imagem {i+1} de {url_imagem}")
        except Exception as e:
            print(f"Erro ao baixar imagem {i+1} de {url_imagem}: {e}")

# tirando a cor das imagens
def tirar_cor(pasta_entrada, pasta_saida_cinza, pasta_saida_binaria):
    if not os.path.exists(pasta_saida_cinza):
        os.makedirs(pasta_saida_cinza)
    if not os.path.exists(pasta_saida_binaria):
        os.makedirs(pasta_saida_binaria)

    for arquivo in os.listdir(pasta_entrada):
        caminho_arquivo = os.path.join(pasta_entrada, arquivo)
        if not caminho_arquivo.lower().endswith(('.jpg', '.jpeg', '.png')):  # Verificar apenas arquivos de imagem
            continue

        print(f"Processando {arquivo}...")

        # Carregar a imagem colorida
        imagem_colorida = np.array(Image.open(caminho_arquivo))

        # Reduzir dimensionalidade
        h, w, c = imagem_colorida.shape
        imagem_cinza = np.zeros((h, w), dtype=np.uint8)
        for i in range(h):
            for j in range(w):
                r, g, b = imagem_colorida[i, j]
                # Média ponderada para tons de cinza
                imagem_cinza[i, j] = int(0.2989 * r + 0.5870 * g + 0.1140 * b)

        # Binario
        imagem_binaria = np.zeros((h, w), dtype=np.uint8)
        for i in range(h):
            for j in range(w):
                imagem_binaria[i, j] = 255 if imagem_cinza[i, j] > 127 else 0

        # Salvar imagens
        Image.fromarray(imagem_cinza).save(os.path.join(pasta_saida_cinza, f"cinza_{arquivo}"))
        Image.fromarray(imagem_binaria).save(os.path.join(pasta_saida_binaria, f"binaria_{arquivo}"))

# Diretórios de entrada e saída
pasta_entrada = "imagens_coloridas"
pasta_saida_cinza = "imagens_cinzas"
pasta_saida_binaria = "imagens_binarias"

# Baixar imagens aleatórias
baixar_imagens(imagens_urls, pasta_entrada, n=10)

# Processar as imagens (convertendo para cinza e binárias)
tirar_cor(pasta_entrada, pasta_saida_cinza, pasta_saida_binaria)
```
